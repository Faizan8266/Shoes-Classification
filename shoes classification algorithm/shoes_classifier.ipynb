```python
from __future__ import print_function  
import cv2
import os
image=[]
l=[]
m=[]
f1=[]
#for boots
for filename in os.listdir("C:/Users/faiza/Anaconda3/boots"):
    img=cv2.imread(os.path.join("C:/Users/faiza/Anaconda3/boots",filename))
    if img is not None:
        image.append(img)
#gray scale
for i in range(0,len(image)):
    j=image[i]
    gray=cv2.cvtColor(j,cv2.COLOR_RGB2GRAY)
    l.append(gray)
#for gladiator
for filename in os.listdir("C:/Users/faiza/Anaconda3/gladiator"):
    img=cv2.imread(os.path.join("C:/Users/faiza/Anaconda3/gladiator",filename))
    if img is not None:
        image.append(img)
#gray scale
for i in range(0,len(image)):
    j=image[i]
    gray1=cv2.cvtColor(j,cv2.COLOR_RGB2GRAY)
    l.append(gray1)
#for sandals
for filename in os.listdir("C:/Users/faiza/Anaconda3/sandals"):
    img=cv2.imread(os.path.join("C:/Users/faiza/Anaconda3/sandals",filename))
    if img is not None:
        image.append(img)
#gray scale
for i in range(0,len(image)):
    j=image[i]
    gray2=cv2.cvtColor(j,cv2.COLOR_RGB2GRAY)
    l.append(gray2)
#for shoes
for filename in os.listdir("C:/Users/faiza/Anaconda3/shoes"):
    img=cv2.imread(os.path.join("C:/Users/faiza/Anaconda3/shoes",filename))
    if img is not None:
        image.append(img)
#gray scale
for i in range(0,len(image)):
    j=image[i]
    gray3=cv2.cvtColor(j,cv2.COLOR_RGB2GRAY)
    l.append(gray3)
#for slippers
for filename in os.listdir("C:/Users/faiza/Anaconda3/slippers"):
    img=cv2.imread(os.path.join("C:/Users/faiza/Anaconda3/slippers",filename))
    if img is not None:
        image.append(img)
#gray scale
for i in range(0,len(image)):
    j=image[i]
    gray4=cv2.cvtColor(j,cv2.COLOR_RGB2GRAY)
    l.append(gray4)
#for sports
for filename in os.listdir("C:/Users/faiza/Anaconda3/sports"):
    img=cv2.imread(os.path.join("C:/Users/faiza/Anaconda3/sports",filename))
    if img is not None:
        image.append(img)
#gray scale
for i in range(0,len(image)):
    j=image[i]
    gray5=cv2.cvtColor(j,cv2.COLOR_RGB2GRAY)
    l.append(gray5)

#resize
for i in range(0,len(l)):
    p=l[i]
    k=cv2.resize(p,(300,300))
    m.append(k)
for i in range(0,len(l)):
    n=[]
    n=m[i].flatten()
    f1.append(n)  

f1_train=[]
for i in range(0,len(l)):
    if(i<=79):
        f1_train.append(0)
    if(i>=80)and(i<=159):
        f1_train.append(1)
    if(i>=160)and(i<=239):
        f1_train.append(2)
    if(i>=240)and(i<=319):
        f1_train.append(3)
    if(i>=320)and(i<=399):
        f1_train.append(4)
    if(i>=400):
        f1_train.append(5)
        
        
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(f1,f1_train,test_size=0.20)


#knn
print('KNN')
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6)
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
pred=knn.predict(x_train)
pred.shape
c=0
for i in range(0,len(y_train)):
    if(pred[i]==y_train[i]):
        c+=1
acc=(float(c)/len(y_train))*100
print(acc)
```
